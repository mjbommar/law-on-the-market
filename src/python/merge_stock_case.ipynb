```python
# Read SCDB data
scdb_data = pandas.read_csv(os.path.join(base_path, "data", "scotus", "SCDB_2016_01_justiceCentered_Citation.csv"), 
                            encoding="iso-8859-1", low_memory=False)
scdb_data.shape
```

```python
ticker_case_data = []

# Iterate through cases
for case_id, case_data in lotm_case_name_df.iterrows():
    # Get matching SCDB data
    scdb_case_data = scdb_data.loc[scdb_data.loc[:, "docket"] == case_data["docket"], :]
    if scdb_case_data.shape[0] == 0:
        print("Unable to match SCDB data for {0}: {1}".format(case_data["docket"], case_data["caseName"]))
        continue
    else:
        vote_margin = scdb_case_data.iloc[0]["majVotes"] - scdb_case_data.iloc[0]["minVotes"]
        case_issue = scdb_case_data.iloc[0]["issue"]
        case_issue_area = scdb_case_data.iloc[0]["issueArea"]
        case_petitioner = scdb_case_data.iloc[0]["petitioner"]
        case_respondent = scdb_case_data.iloc[0]["respondent"]
        law_type = scdb_case_data.iloc[0]["lawType"]
        lc_disagreement = scdb_case_data.iloc[0]["lcDisagreement"]
        case_disposition = scdb_case_data.iloc[0]["caseDisposition"]
    
    # Iterate through all case keys
    for k in case_map.keys():
        # Check for partial key match
        key_tokens = k.split(".")
        if key_tokens[0].lower() == str(case_id).lower():
            # Get key pieces
            if len(key_tokens) > 1:
                test_part = key_tokens[1]
            else:
                test_part = None
            test_name = key_tokens[0]

            # Build records
            for ticker in case_map[k]:
                ticker_case_row = case_data.to_dict()
                ticker_case_row["case"] = case_id
                ticker_case_row["ticker"] = ticker
                ticker_case_row["testPart"] = test_part
                ticker_case_row["voteMargin"] = vote_margin
                ticker_case_row["lowerCourtDisagreement"] = lc_disagreement
                ticker_case_row["caseDisposition"] = case_disposition
                ticker_case_row["lawType"] = law_type                
                ticker_case_row["issue"] = case_issue
                ticker_case_row["issueArea"] = case_issue_area
                ticker_case_row["petitioner"] = case_petitioner
                ticker_case_row["respondent"] = case_respondent
                ticker_case_data.append(ticker_case_row)

ticker_case_df = pandas.DataFrame(ticker_case_data)
ticker_case_df.to_csv(os.path.join(base_path, "data", "events", "case_ticker_data.csv"), index=False)
```
